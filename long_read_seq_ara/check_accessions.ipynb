```python
parallelism = pd.read_csv('../../gea_grene-net/key_files/generation_1_parallelism.txt', sep = '\t')
```

```python
parallelism = parallelism[parallelism['source'] == 'ecotype']
```

```python
h_paral = parallelism.sort_values('mean', ascending=False).head(10)['site'].values
```

```python
e_freq = pd.read_csv('../../gea_grene-net/key_files/merged_ecotype_frequency.txt', sep = '\t')
```

```python
e_freq
```

```python
# select columns that match the condition
selected_cols = [col for col in e_freq.columns 
                 if any(col.startswith(f"{i}_1_") for i in h_paral)]

# subset the DataFrame
e_freq_hp = e_freq[selected_cols]
```

```python
fe_g = pd.read_csv('../../gea_grene-net/key_files/founder_ecotype_names.csv')
```

```python
e_freq_hp.index = fe_g['0'].values
```

```python
e_freq_hp
```

```python
from collections import defaultdict
```

```python
# Group columns by prefix before first '_'
grouped_cols = defaultdict(list)
for col in e_freq_hp.columns:
    group_key = col.split('_')[0]
    grouped_cols[group_key].append(col)
```

```python
e_freq_hp.sort_values('4_1_3',ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize


all_top10_values = []
for cols in grouped_cols.values():
    for col in cols:
        all_top10_values.extend(e_freq_hp[col].nlargest(10).values)

vmin = min(all_top10_values)
vmax = max(all_top10_values)


for group_key, cols in grouped_cols.items():
    fig, axes = plt.subplots(1, len(cols), figsize=(1 * len(cols), 3))

    if len(cols) == 1:
        axes = [axes]

    for ax, col in zip(axes, cols):
        top10_df = e_freq_hp[col].sort_values(ascending=False).head(10)
        highlight = top10_df.index.isin(ecotypes_pangenome)

        sns.heatmap(top10_df.to_frame(), cmap='viridis', square=True,annot=True, fmt=".2f", annot_kws={"size": 7},
                    cbar=False, vmin=vmin, vmax=vmax,
                    yticklabels=top10_df.index,
                    xticklabels=False,
                    
                    ax=ax)

        for j, is_highlighted in enumerate(highlight):
            if is_highlighted:
                ax.add_patch(plt.Rectangle((0, j), 1, 1, 
                                           fill=False, edgecolor='red', linewidth=2))

        ax.set_title(col, fontsize=10)
        ax.tick_params(labelsize=7)

        norm = Normalize(vmin=vmin, vmax=vmax)
    sm = ScalarMappable(norm=norm, cmap='viridis')
    sm.set_array([])  # Required for matplotlib < 3.1

    #cbar = fig.colorbar(sm, ax=axes, orientation='vertical', fraction=0.02, pad=0.01)
    #cbar.set_label('Frequency', rotation=270, labelpad=15)

    plt.tight_layout()
    plt.suptitle(f"Site {group_key} (Red = Pangenome)", fontsize=10, y=1.05)
    plt.savefig(f'site_{group_key}_pg_present.png')
    plt.show()
```

```python
top10_ecotypes_per_site = {}

for col in e_freq_hp.columns:
    # Get top 10 indices for the column
    top_indices = e_freq_hp[col].nlargest(10).index.tolist()
    # Store them in the dictionary
    top10_ecotypes_per_site[col] = top_indices
```
