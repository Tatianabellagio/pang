```python
import pandas as pd
import re

# Path to your file
fname = "gam_stats.txt"

stats = {}

with open(fname) as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        
        # Example: "Total alignments: 6688662"
        if ":" in line:
            key, value = line.split(":", 1)
            key = key.strip()
            value = value.strip()
            
            # extract the first number in value
            # (works even for lines with multiple numbers)
            m = re.search(r"[-+]?\d*\.\d+|\d+", value)
            if m:
                stats[key] = m.group()
            else:
                stats[key] = value

```

```python
import pandas as pd

def compute_metrics(stats):
    # convert values to numeric (they come in as strings)
    s = {k: float(v) for k, v in stats.items()}

    metrics = {
        "total_reads": s["Total alignments"],
        "aligned_reads": s["Total aligned"],
        "aligned_pct": s["Total aligned"] / s["Total alignments"],
        "perfect_reads": s["Total perfect"],
        "perfect_pct": s["Total perfect"] / s["Total alignments"],
        "properly_paired_pct": s["Total properly paired"] / s["Total paired"],
        "mean_mapq": s["Mapping quality"],
        "mean_alignment_score": s["Alignment score"],
        "edit_rate": (s["Insertions"] + s["Deletions"] + s["Substitutions"]) / s["Matches"],
        "softclip_rate": s["Softclips"] / s["Matches"],
    }

    return pd.DataFrame([metrics])

df_metrics = compute_metrics(stats)
df_metrics
```
