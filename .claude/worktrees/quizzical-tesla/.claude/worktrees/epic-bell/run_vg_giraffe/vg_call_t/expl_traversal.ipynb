```python
import pandas as pd

records = []

with open("MLFH041020180321.traversals.gaf") as f:
    for line in f:
        if line.startswith("@"):
            continue
        fields = line.strip().split("\t")

        name = fields[0]
        path = fields[5]
        length = int(fields[6])
        support = int(fields[12].split(":")[-1])

        parts = name.split("#")
        chrom = parts[2]
        pos = int(parts[3])
        snarl = parts[4]
        allele = int(parts[6])

        records.append({
            "chrom": chrom,
            "pos": pos,
            "snarl": snarl,
            "allele": allele,
            "length": length,
            "support": support,
            "path": path,
        })

df = pd.DataFrame(records)
df
```
