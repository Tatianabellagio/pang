```python
import requests
from bs4 import BeautifulSoup

def get_ecotype_from_cs(cs_number):
    url = f"https://arabidopsis.info/StockInfo?accession={cs_number}"
    response = requests.get(url)

    if response.status_code != 200:
        return f"Error: Status code {response.status_code}"

    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the ecotype or accession name on the page
    table_rows = soup.find_all('tr')
    for row in table_rows:
        cols = row.find_all('td')
        if len(cols) == 2 and 'Ecotype' in cols[0].text:
            return cols[1].text.strip()

    return "Ecotype not found"

# Example
cs_number = "CS76945"
ecotype_name = get_ecotype_from_cs(cs_number)
print(f"Ecotype for {cs_number}: {ecotype_name}")
```

```python
import pandas as pd

# Load the file with proper column names
colnames = [
    'internal_id', 'seq_provider', 'name', 'country', 'short_name',
    'latitude', 'longitude', 'collected_by', 'date_collected', 'cs_number',
    'ancestry', 'unknown1', 'unknown2'
]

df = pd.read_csv('../../ara_keyfiles/1001g_accessions.txt', header=None, names=colnames)

# Strip and standardize the CS numbers
df['cs_number'] = df['cs_number'].astype(str).str.strip()

# Function to get ecotype name from CS number
def get_ecotype_from_cs(cs_number):
    match = df[df['cs_number'] == cs_number]
    if not match.empty:
        return match.iloc[0]['name']
    else:
        return "Ecotype not found"

# Example usage
cs_number = 'CS76945'
ecotype_name = get_ecotype_from_cs(cs_number)
print(f"Ecotype for {cs_number}: {ecotype_name}")
```
