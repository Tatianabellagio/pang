```python
import pandas as pd
from pathlib import Path
import numpy as np

BASE = Path("/home/tbellagio/scratch/pang/test_freqk/run_set_02_rep1_S1/")
```

```python
# --- index (chrom/pos) ---
idx_path = BASE / f"set_02_rep1.k31.freqk.ref_index"
idx = pd.read_csv(idx_path, header=None)[[1,2]]

idx.columns = ["chrom", "pos"]
idx["chrom"] = idx["chrom"].astype(str)
idx["pos"] = idx["pos"].astype(int) + 2   # keep your +2 correction
af_path = BASE / "S1.allele_frequencies.k31.dedup.tsv"

rows = []
max_cols = 0

with open(af_path, "r") as f:
    for line in f:
        parts = line.rstrip("\n").split("|")
        rows.append(parts)
        if len(parts) > max_cols:
            max_cols = len(parts)

# pad
rows = [r + [np.nan] * (max_cols - len(r)) for r in rows]

af = pd.DataFrame(rows, columns=[f"c{i}" for i in range(max_cols)])

af.columns = ['ref', 'alt1', 'alt2']

df = pd.concat([idx.reset_index(drop=True),
                     af],
                    axis=1)
```

```python
import pysam
import re
from collections import Counter

vcf = pysam.VariantFile("/home/tbellagio/scratch/pang/test_postproc/set_10_rep3.vcf.gz")

bad = 0
total_multi = 0
examples = []

for rec in vcf:
    k = len(rec.alts or [])
    if k < 2:
        continue
    total_multi += 1

    used = [False] * (k + 1)  # used[j] for j=1..k

    for s in rec.samples.values():
        gt = s.get("GT", None)
        if gt is None:
            continue
        # gt is a tuple like (0,2) or (1,) or (None,None)
        for a in gt:
            if a is None:
                continue
            if 1 <= a <= k:
                used[a] = True

    if not all(used[1:]):
        bad += 1
        if len(examples) < 10:
            missing = [j for j in range(1, k+1) if not used[j]]
            examples.append((rec.chrom, rec.pos, rec.id, k, missing))

print("multiallelic sites:", total_multi)
print("sites with any unused ALT:", bad)
print("example bad sites:", examples)
```
