```python
VCF=Abd-0.tair10.syri.vcf
OUT=Abd-0.syri.SVonly.alt_counts.csv

# Count ALT alleles and write CSV: alt,count
bcftools query -f '%ALT\n' "$VCF" \
  | sort \
  | uniq -c \
  | awk 'BEGIN{OFS=","; print "alt","count"} {print $2,$1}' \
  > "$OUT"

echo "Wrote: $OUT"
column -s, -t "$OUT"
Wrote: Abd-0.syri.SVonly.alt_counts.csv
```
